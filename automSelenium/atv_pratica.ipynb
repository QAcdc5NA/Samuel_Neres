```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get(DEMO_URL)
print("Título da página:", driver.title)
```

#### Execução automatizada — **CT-01** (teste positivo)

```python
# CT-01: aguardar até o formulário existir
wait = WebDriverWait(driver, 10)

# Localização dos elementos pelo id do HTML.
campo_nome = wait.until(EC.presence_of_element_located((By.ID, "nome")))
campo_email = driver.find_element(By.ID, "email")
campo_valor = driver.find_element(By.ID, "valor")
botao_enviar = driver.find_element(By.ID, "submitBtn")

# Interação: limpar e enviar texto como o usuário faria no teclado.
campo_nome.clear()
campo_nome.send_keys("Samuel")

campo_email.clear()
campo_email.send_keys("Samuel@email.com")

campo_valor.clear()
campo_valor.send_keys("500")

# Dispara a validação JavaScript do formulário.
botao_enviar.click()

# Espera explícita até a mensagem ficar visível e leitura do resultado.
mensagem = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
print("Mensagem retornada:", mensagem)

# Validação automatizada (resultado esperado da CT-01).
assert "Sucesso" in mensagem
assert "Samuel" in mensagem
assert "R$" in mensagem

print("Teste positivo (CT-01) executado com sucesso.")
```

#### Execução automatizada — **CT-02** (teste negativo)

```python
# CT-02: volta ao estado inicial da página para não acumular dados da CT-01.
driver.get(DEMO_URL)

wait = WebDriverWait(driver, 10)

campo_nome = wait.until(EC.presence_of_element_located((By.ID, "nome")))
campo_email = driver.find_element(By.ID, "email")
campo_valor = driver.find_element(By.ID, "valor")
botao_enviar = driver.find_element(By.ID, "submitBtn")

campo_nome.clear()
campo_nome.send_keys("Aluno QA")

campo_email.clear()
campo_email.send_keys("email_invalido")  # propositalmente sem '@' para forçar erro

campo_valor.clear()
campo_valor.send_keys("200")

botao_enviar.click()

mensagem = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
print("Mensagem retornada:", mensagem)

# Resultado esperado CT-02: e-mail inválido.
assert "Erro" in mensagem
assert "e-mail inválido" in mensagem

print("Teste negativo (CT-02) executado com sucesso.")
```

#### Execução automatizada — **CT-03 a CT-07** (Limpar, obrigatórios, nome curto, valor inválido)


```python
# CT-03 a CT-06: cada bloco recarrega a demo para estado independente.

wait = WebDriverWait(driver, 10)


def recarregar_demo():
    driver.get(DEMO_URL)
    wait.until(EC.presence_of_element_located((By.ID, "nome")))


# CT-03: botão Limpar esvazia campos e oculta feedback
recarregar_demo()
campo_nome = driver.find_element(By.ID, "nome")
campo_email = driver.find_element(By.ID, "email")
campo_valor = driver.find_element(By.ID, "valor")
driver.find_element(By.ID, "nome").send_keys("Teste")
driver.find_element(By.ID, "email").send_keys("t@t.com")
driver.find_element(By.ID, "valor").send_keys("99")
driver.find_element(By.ID, "btnLimpar").click()

assert campo_nome.get_attribute("value") == ""
assert campo_email.get_attribute("value") == ""
assert campo_valor.get_attribute("value") == ""
area = driver.find_element(By.ID, "resultMsg")
assert area.text == ""
assert not area.is_displayed()
print("CT-03 (Limpar) — OK.")

# CT-04: envio sem preencher → obrigatórios 
recarregar_demo()
driver.find_element(By.ID, "submitBtn").click()
msg4 = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
assert "Erro" in msg4
assert "obrigatórios" in msg4
print("CT-04 (campos obrigatórios) — OK.")

# CT-05: nome com 1 caractere 
recarregar_demo()
driver.find_element(By.ID, "nome").send_keys("X")
driver.find_element(By.ID, "email").send_keys("x@y.com")
driver.find_element(By.ID, "valor").send_keys("10")
driver.find_element(By.ID, "submitBtn").click()
msg5 = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
assert "Erro" in msg5
assert "2 caracteres" in msg5
print("CT-05 (nome curto) — OK.")

# CT-06: valor zero
recarregar_demo()
driver.find_element(By.ID, "nome").send_keys("Maria")
driver.find_element(By.ID, "email").send_keys("maria@test.com")
driver.find_element(By.ID, "valor").send_keys("0")
driver.find_element(By.ID, "submitBtn").click()
msg6 = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
assert "Erro" in msg6
assert "maior que zero" in msg6
print("CT-06 (valor não positivo) — OK.")

# CT-07: valor não positivo
recarregar_demo()
driver.find_element(By.ID, "nome").send_keys("Mick")
driver.find_element(By.ID, "email").send_keys("mick@test.com")
driver.find_element(By.ID, "valor").send_keys("-500")
driver.find_element(By.ID, "submitBtn").click()
msg7 = wait.until(EC.visibility_of_element_located((By.ID, "resultMsg"))).text
assert "Erro" in msg7
assert "maior que zero" in msg7
print("CT-07 (valor menor que zero) — OK.")


print("Todos os testes CT-03 a CT-06 concluídos com sucesso.")
```
